```python
#function to sequence the samples (sequence of 10 instances = 100ms length)
def stack_dataframes(df, Tob, Tp):
    columns = df.shape[1]
    stacked_arrays = []
    btp = []
    #y = []

    for j in range(columns-1):
        x = []
        for i in range(df.shape[0] - Tp - Tob  + 1):
            x.append(df.iloc[i:i+Tob, j])
        x = np.array(x)
        stacked_arrays.append(x)

    stacked_arrays = np.stack(stacked_arrays, axis=2)
    print("X ", stacked_arrays.shape)



    for i in range(x.shape[0]):
        count1 = 0
        for value in df.iloc[Tob + i +1 : Tob + i +1 + Tp, -1]:
          if (value==1):
            count1 =+ 1
        if (count1) : btp.append(1)
        else : btp.append(0)

    btp = np.array(btp).reshape(-1, 1)
    print("btp ", btp.shape)
    return stacked_arrays, btp
```

```python
x1, y1 = stack_dataframes(df1, 10, 40)
x2, y2 = stack_dataframes(df6, 10, 40)
```

```python
x = np.append(x1, x2, axis=0)
y = np.append(y1, y2, axis=0)
print(x.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print(X_train.shape)
print(X_test.shape)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
from tensorflow.keras import optimizers
from tensorflow.keras.activations import linear, sigmoid, relu
```

```python
model = Sequential()
model.add(LSTM(64, return_sequences= True , input_shape=(X_train.shape[1] , X_train.shape[2]), activation='sigmoid'))
#model.add(LSTM(64, return_sequences= True, activation='sigmoid'))
#model.add(LSTM(64, return_sequences= True, activation='sigmoid'))
#model.add(LSTM(64, return_sequences= True, activation='sigmoid'))
model.add(LSTM(64, return_sequences= False, activation='sigmoid'))
model.add(Dropout(0.2))
model.add(Dense(64, activation='linear'))
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping

filepath = '/content/drive/MyDrive/Capstone/RITIKA/models/LSTM_Trajectory1.hdf5'

callbacks = [ModelCheckpoint(filepath, monitor='loss', save_best_only=True, mode='min')] #EarlyStopping(monitor='val_loss', patience=20),
optimizers.SGD(momentum=0.9)
model.compile(optimizer='adam',loss='mse',metrics=['accuracy'])   #loss= 'binary_crossentropy' #SGD = adam
```

```python

history = model.fit(X_train,y_train, validation_split=0.1, epochs=1, verbose=1, callbacks=callbacks) #, batch_size=16

loss, accuracy = model.evaluate(X_test,y_test)
print("Test Loss:", loss)
print("Test Accuracy:", accuracy)
```
